The visualization here is a random graph generated using the Erdos-Renyi model as implemented in NetworkX. In this case, a graph consists of 20 nodes and edge exists between any two nodes with a probability of 50%. The degree centrality is calculated for each node present in the random graph (a metric defining how connected a given node is) and these values are then used for coloring this nodes. Centrality in this context refers to the position of a node relative to the other nodes/vertices in the network. The darker the shade of the node, the higher the centrality value and vice versa using a colormap which in this case is tonic. With regards of matplotlib, the audience can see the nodes as well as their labels (if any) along with how the nodes are connected to each other in the network.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Creating the random graph
rg = nx.erdos_renyi_graph(n=20, p=0.5, seed=42)

# calculating the degree centrality
dc = nx.degree_centrality(rg)

nc = [dc[node] for node in rg.nodes]

# visualizing the networks
plt.figure(figsize=(10, 8))  
pos = nx.spring_layout(rg, seed=42)  

nx.draw_networkx(
    rg,
    pos=pos,                       
    node_color=nc,       
    cmap=plt.cm.viridis,           
    with_labels=True,             
    node_size=700,                
    font_size=10,                  
    font_color='white',           
    edge_color='gray',            
    linewidths=1                   
)

plt.title('Random Network Graph with Node Colors Based on Degree Centrality', fontsize=16)
labels = nx.draw_networkx_labels(rg, pos)
plt.axis("off")
plt.show()
```

**Question : 6**

Data visualization requires a careful choice of colors as this directly impacts how the information is perceived and comprehended. Proper use of colors helps achieve clarity, accessibility and cohesion in design. Wrong application of color can mask data trends, derail the audience’ focus or alienate the audience with a visual impairment such as color blindness.

-> **Color Contrast:**
Because of contrast the elements are distinguishable. Readability generally improves with the extremes of colors, for instance when light letters are placed against dark background. 
For example; a line graph with several lines in varying shades of black and white is easy to compare.

-> **Color Harmony**
The use of harmonious color schemes like analogous or complementary contrasts, contributes to the beautification of the visualization. 
For Example: A gradient from blue to green is employed in heat maps to represent the values in a fluid motion. 

-> **Accessibility**
Into account the audience that has some sort of vision related problem such as colors blindness. Refrain from using the inappropriate combinations such as red and green and use colors whose palettes are easy for viewers of all the colors such as Viridis of Cividis.

**Good vs. Bad Color Choices**


```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Category A', 'Category B', 'Category C']
values = [30, 45, 25]

# Good Example
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=['#0072B2', '#56B4E9', '#F0E442'])  # Colorblind-friendly palette
plt.title("Good Example: Accessible Colors", fontsize=16)
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.show()
```

```python
# Bad Example
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=['red', 'green', 'red'])
plt.title("Bad Example: Red-Green Combination", fontsize=16)
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.show()
```

**Color Contrast: High vs. Low**

```python
# High Contrast Example
x = np.linspace(0, 10, 100)
plt.figure(figsize=(8, 5))
plt.plot(x, np.sin(x), label='Group A', color='blue', linewidth=2)
plt.plot(x, np.cos(x), label='Group B', color='orange', linewidth=2)
plt.title("Good Example: High Contrast", fontsize=16)
plt.xlabel("X-Axis", fontsize=12)
plt.ylabel("Y-Axis", fontsize=12)
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

```python
# Low Contrast Example
plt.figure(figsize=(8, 5))
plt.plot(x, np.sin(x), label='Group A', color='#CCCCCC', linewidth=2)  # Light gray
plt.plot(x, np.cos(x), label='Group B', color='#E5E5E5', linewidth=2)  # Lighter gray
plt.title("Bad Example: Low Contrast", fontsize=16)
plt.xlabel("X-Axis", fontsize=12)
plt.ylabel("Y-Axis", fontsize=12)
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

**Colorblind-Friendly Example**

```python
import seaborn as sns

# Load a sample dataset
tips = sns.load_dataset('tips')

# Scatter plot using a colorblind-friendly palette
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', palette='colorblind', s=100)
plt.title("Colorblind-Friendly Scatter Plot", fontsize=16)
plt.xlabel("Total Bill", fontsize=12)
plt.ylabel("Tip", fontsize=12)
plt.legend(title="Day")
plt.show()
```

```python
# Gradient Example using Viridis
heatmap_data = np.random.rand(10, 10)

plt.figure(figsize=(8, 6))
plt.imshow(heatmap_data, cmap='viridis')
plt.colorbar(label="Value")
plt.title("Good Example: Accessible Heatmap (Viridis)", fontsize=16)
plt.show()
```

The process of data visualization is the selection of colors, as it determines what and how this information will be interpreted by people. The right choice of color brings a sharper sight, more readable text and a better interest to the viewer while inappropriate optic combinations distort the perception, create confusion or even segregation of certain groups of people, especially those with sight disabilities such as color blindness. In color theory there contrast, harmony and inclusiveness are predominant. Color patterns with high contrast allow for visual separation of objects in the figure or visualization whereas in low contrast, there is likelihood of visual merge defining an increase in legibility issues. Some examples of these are analogous or complementary such palettes in which such colors that are used are very attractive and capture the attention of a viewer to the desired spot. On the other hand, colors that are not complementary can be an eyesore and take away the focus from what is really important.

Making sure there is accessibility assesses color design because distance collated studies show that colour blindness affects approximately 8% of men and 0.5% of women especially the red green variant which is the most common. In case of inclusivity, red green combinations must be evaded and color blind friendly palettes such as Viridis or Cividis applied. Putting colors limit patterns or textures, or even labels assist in this course. For instance, instead of using a rainbow or a cool gradient that is misleading, using a gradient like the Viridis for heat maps ensures perceptual uniformity and even considers the audience with color vision deficiency.

Appropriate and effective use of colors can be noted in bar graphs or scatter plots that employ distinct and unique colors and allow easy distinction of any categories or data points represented. On the other hand, poorly designed examples such as those that mix red and green colors, or designs that use colors with low apparent contrast, tend to hide the relationships between the data and make it hard to give a meaningful interpretation. Features such as the color-deficient-ready color palette of Seaborn, and the colormap abilities of Matplotlib allow constructing good visuals much easier. Testing visualizations in color, for example, with Color oracle or simulators such as Coblis can help designers ensure the effectiveness of the visuals for everyone, including those who cannot see the full range of colors. In this way, and by maintaining the principles of contrast, harmony, and accessibility, it is possible to use data visualizations as effective means of communication.

Seaborn helps in visualizing the Titanic data set, which consists of numerous numerical and categorical attributes such as age, fare, class,sex, etc. While creating a correlation heatmap, numeric only columns are chosen by select_dtypes(include='number') because correlation calculations cannot be performed for non-numeric variables such as sex or class. The pairwise correlations of those numeric columns are computed using the .corr() method, and the resulting relationships, as exhibited in the heatmap, contain annotations in correlation of this heatmap, indicating their strength and direction respectively. After the heatmap, a scatter plot is drawn to show passenger age against the fare paid, and different ticket classes are in colors. This helps in making an overarching view of the relationships of all the features in a heatmap, while also allowing for close examination of certain data points and clusters in the scatter plot. Such combinations of visualizations enhance the comprehension of trends and outliers in Titanic data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Step 2: Filter Numeric Columns
# Keep only numeric columns for correlation
numeric_columns = titanic.select_dtypes(include='number')
correlation_matrix = numeric_columns.corr()  # Compute correlations for numerical columns

# Step 3: Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Titanic Dataset', fontsize=16)
plt.show()

# Step 4: Scatter Plot of Age vs. Fare
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    titanic['age'], 
    titanic['fare'], 
    c=titanic['class'].map({'First': 'blue', 'Second': 'green', 'Third': 'red'}), 
    alpha=0.7
)
plt.title('Age vs. Fare Scatter Plot', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Fare', fontsize=12)
plt.grid(alpha=0.5)
plt.legend(['First Class', 'Second Class', 'Third Class'], loc='upper right')
plt.show()
```
